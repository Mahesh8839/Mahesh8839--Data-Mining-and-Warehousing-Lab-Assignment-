```python
#Importing libraries
import numpy as np
import pandas as pd
from openpyxl import Workbook
```

```python
#Reading given dataset
data=pd.read_csv("jio_mart_items.csv")
```

```python
data.head()  #Display first 5 rows
```

```python
data.info()
```

```python
new_data=data[['category',	'sub_category',	'href',	'items',	'price']] # features of dataset
new_data.head()
```

```python
new_data.isnull().sum()
```

```python
new_data[data['items'].isna()]
new_data[data['price'].isna()]
```

```python
new_data['price']=new_data['price'].fillna(new_data['price'].mean())
new_data.head()
```

```python
new_data.isnull().sum()
```

```python
new_data['items']=new_data['items'].fillna(new_data['items'].mode()[0])
new_data.head()
```

```python
print(new_data['items'].mode()[0])
```

```python
new_data.isnull().sum()
```

```python
new_data.info()
```

```python
set1=data[['category',	'sub_category',	'href']] #Data Reduction
set1.head()
```

```python
 #Data Reduction
set2=data[['items','price']] #Data Reduction
set2.head()
```

```python
display(set1,set2)
```

```python
#Merge dataset
set3=pd.merge(new_data,set2, left_index=True, right_index=True)
set3.head(25)
```
